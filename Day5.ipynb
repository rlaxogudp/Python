```python
# 예외 처리
# 코드 작성 시 발생할 수 있는 예기치 않은 상황에 대해 개발자가 처리함으로써 
# 코드의 안정성과 신뢰성을 높이는 문법
# try : 오류가 발생할 수 있는 코드 
# except : 오류를 처리하는 코드
# finally : 반드시 실행되어야 하는 코드
try:
    number = int(input("자연수를 입력하세요 :"))
    result = 10 / number
    print(f"10에서 나눈값 : {result}")
except ZeroDivisionError:
    print("0으로 나눌 수 없습니다")
except ValueError:
    print("자연수를 입력해주세요")
finally: # 어떤 상황이든 실행된다
    print("프로그램 종료")
```

```python
with open("train.csv", "r") as file:
    reader = csv.DictReader(file)
    male_survived = 0
    male_passed = 0
    female_survived = 0
    female_passed = 0
    for r in reader:
        if int(r["Survived"]) == 1 and str(r["Sex"]) == "male":
            male_survived += 1
        elif int(r["Survived"]) == 0 and str(r["Sex"]) == "male":
            male_passed += 1
        elif int(r["Survived"]) == 1 and str(r["Sex"]) == "female":
            female_survived += 1
        elif int(r["Survived"]) == 0 and str(r["Sex"]) == "female":
            female_passed += 1
    total = male_survived + male_passed + female_passed + female_survived
    # 남성의 생존 확률
    #여성의 생존 확률
    per_male = male_survived/(male_passed + male_passed) * 100
    per_femail = female_survived/(female_passed + female_survived) * 100
    print(f"생존한 남성 : {male_survived}")
    print(f"사망한 남성 : {male_passed}")
    print(f"생존한 여성 : {female_survived}")
    print(f"사망한 여성 : {female_passed}")
    print(f"전체 승객의 수 : {total}")
    print(f"남성의 생존확률 : {per_male}%")
    print(f"여성의 생존확률 : {per_femail}%")
```

```python
# Json(JavaScrip Object Notation)

# 영화 진흥 위원회 API를 통해 2025.07.06 Json을 가져오자

!curl -L https://github.com/nayeho/harman2/raw/refs/heads/main/movie.json -o movie.json
```

```python
import json
# movie.json -> 2025.07.06 일별 박스오피스 순위
data = ""
with open("movie.json", "r", encoding="utf-8") as file:
    data = json.load(file)
    boxOfficeResult = data["boxOfficeResult"]
    dailyBoxOfficeList = boxOfficeResult["dailyBoxOfficeList"]

    total_audi = 0
    for i in range(len(dailyBoxOfficeList)):
        rank = dailyBoxOfficeList[i]["rank"]
        movieNM = dailyBoxOfficeList[i]["movieNm"]
        audiCnt = dailyBoxOfficeList[i]["audiCnt"]
        total_audi += int(audiCnt)
        print(f"{rank}위 : {movieNM} ({audiCnt}명)")
    print(f"총 관객 : {total_audi}")
```
