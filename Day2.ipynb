```python
# 논리 연산자
print(not True)
print(not False)
print(not (10 > 5))
# 논리 곱(and)
# 1 -> True
# 0 -> False
# 1 x 1 = 1 (True and True) -> True
# 1 x 0 = 0 (True and False) -> False
# 0 x 1 = 0 (False and True) -> False
# 0 x 0 = 0 (False and False) -> False
print((10 < 5) and (3 > 1)) 
# and 앞쪽에서 이미 False이기 때문에
# and 뒷쪽은 Dead Code가 된다!! (연산 X)
# 논리 합(or)
# 1 + 1 = 2 (True or True) -> True
# 1 + 0 = 1 (True or False) -> True
# 0 + 1 = 1 (False or True) -> True
# 0 + 0 = 0 (False or False) -> False
print((10 > 5) or (3 > 1)) # True
# or 앞쪽에서 이미 True이기 때문에
# or 뒷쪽은 Dead Code가 된다!! (연산 X)
```

```python
# 사용자로부터 자연수를 입력 받아
# 지속적으로 누적 합하여 출력하는 프로그램

add_total = 0
# 사용자가  0을 입력하면 종료

while user_input != 0:
    user_input = int(input("자연수를 입력 :"))
    add_total += user_input
    print("누적 합 :", add_total)
```

```python
##### 다이어트 프로그램
# 1. 현재 몸무게 입력: 
# current_weight
# 2. 목표 몸무게 입력:
# goal_weight

current_weight = int(input("현재 몸무게:"))
goal_weight = int(input("목표 몸무게:"))
week = 1

while(1):
    weight = int(input("%d주차 몸무게 :"%week))
    current_weight -= weight
    week = week + 1
    if current_weight <= goal_weight :
        print("%dKG달성! 축하합니다."%current_weight)
        break
```

```python
# 1~100까지 수 중 하나를 맞추는 게임
# random 랜덤관련 라이브러리

import random
answer = random.randint(1,100)
#print(answer)
```

```python
# 숫자를 입력하세요(1~100)
number = int(input("숫자를 입력하세요:"))

while(1):
    if number < answer :
        print("더 크다")
        number = int(input("숫자를 다시 입력하세요:"))
    elif number > answer :
        print("더 작다")
        number = int(input("숫자를 다시 입력하세요:"))
    elif number == answer :
        print("정답")
        break
```
