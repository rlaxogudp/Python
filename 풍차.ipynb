```python
num = int(input("풍차 날개 길이 입력 :"))
for i in range(1, num + 1):
    for j in range(i):
        print("*", end="")
    for j in range(num-1-i,-1,-1):
        print(" ", end="")
    for j in range(num-i,-1,-1):
        print("*", end="")
    print()
for i in range(1, num+1):
    for j in range(num-1-i,-1,-1):
        print(" ", end= "")
    for j in range(i):
        print("*", end="")
    for j in range(i-1):
        print(" ", end="")
    for j in range(num-i,-1,-1):
        print("*", end="")
    print()

import time
time.sleep(5)
```
