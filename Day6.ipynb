```python
# 조건을 통한 데이터 추출
# 슬라이싱
# df명.loc[행 슬라이싱]
df_titanic.loc[0:5] #0 <= data <= 5
# df명.loc[행 슬라이싱, 열 슬라이싱]
df_titanic.loc[0:5, "Survived":"Age"]

df_titanic.loc[100:103, "Pclass":"Sex"]
```

```python
# Boolean Indexing
# True, False를 활용하여 데이터를 추출
# df의 Age 컬럼의 데이터가 30이상 (Series)
df_titanic["Age"] >= 30
df_titanic.loc[df_titanic["Age"] >= 30]
df_titanic["Survived"] == 1
df_titanic.loc[(df_titanic["Survived"] == 1) & (df_titanic["Age"] >= 30)]
df_titanic.loc[(df_titanic["Survived"] == 1) & (df_titanic["Sex"] == "male")]
```

```python
# iloc(integer location)
# 행과 열의 인덱스 정수를 활용하여 데이터 추출
# df명.iloc[행 인덱스 정수, 열 인덱스 정수]
df_titanic.iloc[0, 3] # 첫 번째 사람의 4번째 컬럼에 해당하는 데이터
# 인덱스 정수 -> for문 등 순차적 index를 활용하는 일련의 행위 가능
# iloc 슬라이싱
# df명.iloc[행 인덱스 정수 슬라이싱]
# df명.iloc[행 슬라이싱, 열 슬라이싱]
df_titanic.iloc[0:3] # 0 <= data < 3
df_titanic.iloc[0:3, 1:4]
# iloc Boolean indexing X
# for -> index를 꼭 활용 -> iloc
# Bloolean indexing을 꼭 써야한다 -> loc

# describe() -> 통계적 정보를 제공
```

```python
# Lambda Expression
# Lambda Funtion
# 부서 데이터를 소문자로
def lower_department(department):
    return department.lower()
# 함수명 + 매개변수 + 리턴값
# (lanbda 매개변수 : 리턴값)(입력값)  1회성 함수
df_copy2["Department"] = df_copy2["Department"].apply(lambda department : department.lower())
df_copy2
```

```python
df_copy2["Salary"] = df_copy2["Salary"].apply(lambda x : x*2)
df_copy2
```

```python
# 데이터 변환
# Name 컬럼의 모든 이름을 대문자로
# .apply (사용자 정의 함수)
def upper_name(name):
    return name.upper()

df_copy2["Name"] = df_copy2["Name"].apply(upper_name)
df_copy2

def plus_500_salary(salary):
    return salary + 500
df_copy2["Salary"] = df_copy["Salary"].apply(plus_500_salary)
df_copy2
```

```python
# 간단한 변경하는 map
department_map = {"hr":"Human Resource",
                 "it":"Information Technology",
                 "finance":"Finance"}

df_copy2["Department"] = df_copy2["Department"].map(department_map)
df_copy2
```
