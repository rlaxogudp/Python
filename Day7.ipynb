```python
# 요청 - 응답
# Client - Server
# 하나의 요청 - 하나의 응답
# GET, POST - byte, html
```

```python
# 요청을 위한 라이브러리
import requests
url_news = "https://news.nate.com/recent?mid=n0105"
response = requests.get(url_news)
# 응답 코드
# 200 : 요청 성공
# 403 : 권한 없음
# 404 : 페이지 없음
# 500 : 서버 문제
```

```python
# 응답을 처리하는 3가지 방법
# 1. 문자열
response.text
# 2. byte (이미지, 파일)
response.content
# 3. json (json타입)
response.json()
```

```python
html_news = response.text

# 정적인 웹 페이지의 태그 등에 접근하여 내용물을 출력, 처리할 수 있는 BeautifulSoup 라이브러리 사용
from bs4 import BeautifulSoup as bs

soup_news = bs(html_news, "html.parser")
# soup.find(태그명)
soup_news.find("h2")
# 모든 h2 태그를 가져온다
soup_news.find_all("h2")[0]

# 가져온 태그.text -> 태그 안에 내용물(content)
soup_news.find_all("h2")[5].text

news_list = soup_news.find_all('h2')
for news in news_list:
    if "AI" in news.text or "SKT" in news.text:
        print(news.text)
```
