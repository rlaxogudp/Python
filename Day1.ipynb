### 연산자(Operate)
- 산술 연산자
- 지수 연산자
- 대입 연산자
- 관계 연산자
- 논리 연산자
- 삼항 연산자

```python
### 산술 연산자
num3 = 10
num4 = 7
print(num3 + num4)
print(num3 - num4)
print(num3 * num4)
print(num3 / num4)
print(num3 // num4)
print(num3 % num4)
```

```python
# 지수 연산자(**)
print(2**10)
```

```python
# 대입 연산자 (=)
# 변수명 = 값
# 연산자의 우선순위
age = 20
age = age + 1
age += 1
# 복합 대입 연산자
print(age)
```

```python
# 초를 입력 받아 시, 분, 초로 환산하는 프로그램
total = int(input("초를 입력하세요:"))

hour = int(total // 3600)
num = int(total % 3600)
minute = int(num // 60)
second = int(num % 60)
print(hour, "시간")           
print(minute, "분")
print(second, "초")
```
