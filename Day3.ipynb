```python
# 버블 정렬(BUbble Sort)
# 두 입접한 원소를 비교하여
# 오름차순/내림차순
```

```python
lotto = [8, 17, 12, 27, 5, 7]
# 오름차순 정렬 [5, 7, 8, 12, 17, 27]

# 두 요소의 위치를 바꾼다 -> 치환
# 메모리에서 치환


# 리스트 요소의 개소 : 6개
# 첫 번째 사이클에서 5번 비교
# 두 번째 사이클에서 4번 비교 ...

for n in range(5):
    if lotto[n] > lotto[n+1]:
        temp = lotto[n]
        lotto[n] = lotto[n+1]
        lotto[n+1] = temp

print("cycle1",lotto)

for n in range(4):
    if lotto[n] > lotto[n+1]:
        temp = lotto[n]
        lotto[n] = lotto[n+1]
        lotto[n+1] = temp
print("cycle2", lotto)

for j in range(5,0,-1):
    for n in range(j):
        if lotto[n] > lotto[n+1]:
            temp = lotto[n]
            lotto[n] = lotto[n+1]
            lotto[n+1] = temp
print("cycle", lotto)
```

```python
# 선택 정렬 (Selection Sort)

array = [98, 7 ,70 ,13, 24]
# 원하는 최종 결과 : [7, 13, 24, 70, 98]

for j in range(len(array),1,-1):
    max_index = 0
    max_value = 0
    for i in range(j):
        if array[i] > max_value:
            max_value = array[i]
            max_index = i
    temp = array[max_index]
    array[max_index] = array[j-1]
    array[j-1] = temp

print(array)
    
```

```python
# Bibary Search (이진 탐색)
# list2 안에 78 찾기
search = int(input())

low_index = 0
high_index = len(list2)
mid_index = (low_index + high_index) // 2

is_find = False # 찾으면 True 못찾으면 False

while low_index <= high_index:
    if search == list2[mid_index]:
        print(f"찾고자하는 값은 {mid_index}에 있습니다")
        is_find = True
        break
    elif search > list2[mid_index]:
        low_index = mid_index + 1
        mid_index = (low_index + high_index) // 2
    elif search < list2[mid_index]:
        high_index = mid_index - 1
        mid_index = (low_index + high_index) // 2
if is_find == False:
    print("찾고자 하는 값이 없습니다")
```

```python
# DocString
# Shift +Tab
# 두 값의 차이를 절댓값으로 리턴
def absolute_value(num1,num2):
    """ 두 값의 차이를 절댓값으로 리턴하는 함수"""
    if num1 > num2:
        return num1 - num2
    else:
        return num2 - num1
```

```python
# 1. 약수 여부를 판별하는 is_divisor(num1,num2)
# 만약 num2가 num1의 약수라면 True 리턴
# 약수가 아니라면 False 리턴

def is_divisor(num1,num2):
    if num1 % num2 == 0:
        return True
    else:
        return False
        
is_divisor(10,3)
```

```python
# 2. 모든 약수를 출력하는 print_divisor(num)

def print_divisor(num):
    print(f"{num}의 약수 :",end="")
    for i in range(1,num+1):
        if num % i == 0:
            print(i, end=" ")
               
print_divisor(100)
```

```python
# 3. 약수의 총합을 리턴하는
# get_som_of_divisors(num)

def get_sum_of_divisors(num):
    sum = 0
    print(f"{num}의 약수의 총합 :",end="")
    for i in range(1,num+1):
        if num % i == 0:
            sum += i
    print(sum)

get_sum_of_divisors(10)
get_sum_of_divisors(1000)
```

```python
# 4. 완전수(Perfect Number)
# 자기 자신을 제외한 약수의 합이 자기 자신과 같을 때
# 6 -> 1 2 3
# 28 -> 1 2 4 7 14
# is_perfect_number(num)

def is_perfect_number(num):
    total = 0
    for i in range(1,num):
        if num % i == 0:
            total += i
    if total == num :
        return True
    else:
        return False
is_perfect_number(7)
```

```python
# 5. is_perfect_number을 사용해 3번째 완전수 구해줘

for i in range(2,10000):
    if is_perfect_number(i):
        print(i)
```
