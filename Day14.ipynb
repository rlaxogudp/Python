```python
with open("and_gate.v", "r") as file:
    verilog_code = file.read()
print(verilog_code)
```

```python
lines = verilog_code.strip().splitlines()

inputs = []
output = ""
expression = ""

for line in lines:
    line = line.strip(";").strip()
    if line.startswith("module"):
        start = line.find("(")
        end = line.find(")")
        parameter_section = line[start+1 : end]
        parameter_list = []
        for parameter in parameter_section.split(","):
            parameter_list.append(parameter.strip())
        for parameter in parameter_list:
            if parameter.startswith("input"):
                inputs.append(parameter.replace("input", "").strip())
            elif parameter.startswith("output"):
                output = parameter.replace("output", "").strip()
            else: 
                if output == "":
                    inputs.append(parameter)
    elif line.startswith("assign"):
        line = line.strip()
        left = line.split("=")[0]
        left = left.strip()
        left = left.replace("assign", "").strip()
        right = line.split("=")[1]
        right = right.strip()
        expression = f"{left.strip()} = {right.strip()}"

print(inputs)
print(output)
print(expression)
```

```python
verilog_func = []
verilog_func.append(f"def and_gate({', '.join(inputs)}):")
verilog_func.append(f"    {expression}")
verilog_func.append(f"    return {output}")
result = '\n'.join(verilog_func)
# print(result)
exec(result)
```

```python
print(and_gate(0,0))
print(and_gate(0,1))
print(and_gate(1,0))
print(and_gate(1,1))
```
