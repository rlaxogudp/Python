```python
# 23. 매니저의 사번 및 그 매니저 밑 사원들 중 최소 급여를 받는 사원의 급여를 출력하시오 - 매니저가 없는 사람들은 제외한다.
# - 최소 급여가 5000 미만인 경우는 제외한다.
# - 급여 기준 역순으로 조회한다.

conn = sqlite3.connect("exam.db")
curs = conn.cursor()

sql = """
SELECT manager_id, MIN(salary)
FROM employees
WHERE manager_id IS NOT NULL
GROUP BY manager_id
HAVING MIN(salary) >= 5000
ORDER BY MIN(salary) DESC
"""

import pandas as pd
df = pd.read_sql(sql,conn)

curs.close()
conn.close()

df
```
