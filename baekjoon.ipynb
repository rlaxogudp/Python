```python
while True:
    a, b = map(int,input().split())
    if a > 0 & a < 10:
        if b > 0 & b < 10:
            print(a + b)
        else:
            break
```

```python
T = int(input())

for _ in range(T):
    A, B = map(int,input().split())
    print(A + B)
```

```python
X = int(input())
N = int(input())
total = 0
for _ in range(N):
    a, b = map(int,input().split())
    total += (a*b)

if total == X:
    print("Yes")
else:
    print("NO")
```
